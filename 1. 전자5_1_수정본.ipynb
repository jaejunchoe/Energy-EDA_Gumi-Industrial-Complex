```python
from sklearn.neighbors import NearestNeighbors
import matplotlib.pyplot as plt

# 계약번호별로 그룹화하여 NaN 개수 계산
nan_counts_per_contract = df.groupby('계약번호').apply(lambda group: group.isnull().sum().sum())

# DBSCAN 입력을 위한 DataFrame 생성
nan_counts_df = pd.DataFrame(nan_counts_per_contract, columns=['NaN_count']).reset_index()

# NearestNeighbors로부터 k-거리 그래프 생성
neighbors = NearestNeighbors(n_neighbors=2)
neighbors_fit = neighbors.fit(nan_counts_df[['NaN_count']])
distances, indices = neighbors_fit.kneighbors(nan_counts_df[['NaN_count']])

# 거리를 정렬된 순서로 정렬
sorted_distances = np.sort(distances[:, 1], axis=0)

# 기울기 계산 (차분을 통해)
gradients = np.diff(sorted_distances)

# 기울기가 가장 큰 지점 찾기
greatest_gradient_index = np.argmax(gradients)

# 해당 지점의 eps 값 추정
estimated_eps = sorted_distances[greatest_gradient_index]

print(f"Estimated eps value: {estimated_eps}")


# k-거리 그래프 그리기
plt.plot(sorted_distances)
plt.xlabel('Points sorted by distance')
plt.ylabel('kth nearest distance')
plt.title('k-Distance Graph')
plt.show()

# 이 그래프를 통해 '무릎'을 찾아 `eps` 값을 추정할 수 있습니다.
# min_samples는 실험적으로 결정해야 할 수 있습니다.
```

```python
from sklearn.cluster import DBSCAN

# '무릎' 지점을 찾아 추정된 eps 값 설정
estimated_eps = 686  # 예시 값

# min_samples 값에 대한 범위 설정
min_samples_options = [2, 3, 4, 5]

# DBSCAN 실행 전 데이터 프레임 예시 생성 (실제 사용 시 데이터 프레임을 로드해주세요)
# nan_counts_df = pd.DataFrame({
#     'NaN_count': [10, 20, 5, 30, 40, 50, 5, 20]  # 예시 데이터
# })

for min_samples in min_samples_options:
    dbscan = DBSCAN(eps=estimated_eps, min_samples=min_samples)
    nan_counts_df['cluster'] = dbscan.fit_predict(nan_counts_df[['NaN_count']])

    # 클러스터링 결과의 요약 정보 출력
    cluster_counts = nan_counts_df['cluster'].value_counts()
    # 노이즈 포인트(-1)를 제외한 클러스터 정보만 필터링
    filtered_cluster_counts = cluster_counts[cluster_counts.index != -1]

    print(f"min_samples: {min_samples}")
    print(f"Number of clusters (excluding noise): {len(filtered_cluster_counts)}")
    print(f"Cluster sizes (excluding noise): \n{filtered_cluster_counts}")
    print("\n")
```

```python
# DBSCAN 실행
# 'NaN_count' 열을 사용하여 클러스터링합니다.
# eps와 min_samples는 상황에 맞게 조정해야 합니다.
dbscan = DBSCAN(eps=686, min_samples=5)
nan_counts_df['cluster'] = dbscan.fit_predict(nan_counts_df[['NaN_count']])

# 클러스터링 결과 출력
print(nan_counts_df)
```

```python
from sklearn.cluster import DBSCAN
import numpy as np
import matplotlib.pyplot as plt

# DBSCAN 클러스터링 실행
#dbscan = DBSCAN(eps=686, min_samples=5)
#nan_counts_df['cluster'] = dbscan.fit_predict(nan_counts_df[['NaN_count']])

# 클러스터링 결과에서 노이즈 포인트를 제외한 고유한 클러스터 레이블을 가져옵니다.
unique_clusters = set(nan_counts_df['cluster']) - {-1}  # 노이즈 포인트 제외
colors = plt.cm.Spectral(np.linspace(0, 1, len(unique_clusters)))

plt.figure(figsize=(12, 8))

for cluster_label, color in zip(unique_clusters, colors):
    # 클러스터에 속하는 데이터 포인트를 찾습니다.
    cluster_data = nan_counts_df[nan_counts_df['cluster'] == cluster_label]
    # 클러스터 데이터 포인트를 플롯에 추가합니다.
    plt.scatter(cluster_data.index, cluster_data['NaN_count'], s=50, color=color, label=f'Cluster {cluster_label}')

plt.title('DBSCAN Clustering of NaN Counts (Excluding Noise Points)')
plt.xlabel('Index')
plt.ylabel('NaN Count')
plt.legend()
plt.show()
```

```python
import pandas as pd

# 'df_1' 데이터프레임이 이미 있다고 가정

def fill_missing_values_grouped(df):
    # 날짜를 datetime으로 변환
    df['날짜'] = pd.to_datetime(df['날짜'])

    # '계약번호'별로 그룹화하여 각 그룹에 대해 누락된 값을 채움
    df_filled = df.groupby('계약번호').apply(lambda group: fill_missing_values(group)).reset_index(drop=True)

    return df_filled

def fill_missing_values(df):
    # NaN 값이 있는 행들을 찾음
    nan_indices = df[df['전력량'].isna()].index

    for idx in nan_indices:
        # 현재 행의 날짜와 시간을 가져옴
        current_date = df.loc[idx, '날짜']
        current_time = df.loc[idx, '시간']

        # 전날과 다음날의 날짜를 계산
        prev_date = current_date - pd.Timedelta(days=1)
        next_date = current_date + pd.Timedelta(days=1)

        # 동일 시간대의 전날과 다음날의 데이터를 가져옴
        prev_day_value = df[(df['날짜'] == prev_date) & (df['시간'] == current_time)]['전력량']
        next_day_value = df[(df['날짜'] == next_date) & (df['시간'] == current_time)]['전력량']

        # 두 값의 평균을 구하고, NaN이면 값을 채우지 않음
        if not prev_day_value.empty and not next_day_value.empty:
            df.loc[idx, '전력량'] = (prev_day_value.iloc[0] + next_day_value.iloc[0]) / 2

    # 람다 함수를 사용하기 때문에 groupby 객체의 인덱스를 유지하지 않기 위해 df 반환
    return df

# 'df_1'에 수정된 함수를 적용
df_1_filled = fill_missing_values_grouped(df_1)

# 결과 확인
print(df_1_filled.head(24))
```

```python
import pandas as pd

# 이 코드는 'df_1'이라고 가정한 데이터프레임에 '전력량' 열에서 NaN 값이 있는 행만 찾아서,
# 해당 행의 '날짜'와 '시간'에 해당하는 다음 5일 간의 데이터의 평균으로 NaN 값을 채웁니다.

def fill_missing_values(df, days=5):
    # '날짜'를 datetime으로 변환하고, '시간'을 int로 변환합니다.
    df['날짜'] = pd.to_datetime(df['날짜'])
    df['시간'] = df['시간'].astype(int)

    # '계약번호', '날짜' 및 '시간'을 기준으로 데이터를 정렬합니다.
    df.sort_values(by=['계약번호', '날짜', '시간'], inplace=True)

    # 각 '계약번호'와 '시간' 그룹에 대해 NaN 값을 채우는 로직을 적용합니다.
    def fill_group(group):
        # NaN 값이 있는 행에 대해서만 처리합니다.
        nan_indices = group[group['전력량'].isna()].index
        for idx in nan_indices:
            # 현재 행의 날짜보다 크고, 같은 '시간'을 가진 행을 필터링합니다.
            future_values = group[(group['날짜'] > group.loc[idx, '날짜']) &
                                  (group['시간'] == group.loc[idx, '시간'])]['전력량']
            # 다음 5일 간의 데이터의 평균으로 NaN 값을 채웁니다.
            if not future_values.empty and len(future_values) >= days:
                df.loc[idx, '전력량'] = future_values.iloc[:days].mean()

        return group

    # '계약번호'와 '시간'을 기준으로 그룹화하고, 각 그룹에 대해 fill_group 함수를 적용합니다.
    df = df.groupby(['계약번호', '시간'], group_keys=False).apply(fill_group)

    return df

# 'df_1' 데이터프레임에 함수를 적용합니다.
df_1_filled1 = fill_missing_values(df_1_filled)

# 채워진 데이터프레임의 정보와 상위 24개 행을 출력합니다.
df_1_filled1.info()
df_1_filled1.head(24)
```

```python
def fill_nan_with_future_mean(df):
    # '날짜' 열이 datetime 타입이 아니라면 변환
    df['날짜'] = pd.to_datetime(df['날짜'], errors='coerce')
    # '시간' 열이 정수 타입이 아니라면 변환 (시간이 '1시' 등의 문자열로 되어 있다면 아래 코드를 적용)
    # df['시간'] = df['시간'].str.extract('(\d+)').astype(int)

    # NaN 값을 채울 새로운 DataFrame 생성
    new_df = df.copy()

    # 각 '계약번호'와 '시간'에 대해 NaN 값을 채우기
    for (contract, hour), group in df.groupby(['계약번호', '시간']):
        # NaN 값의 인덱스를 찾기
        nan_indices = group[group['전력량'].isna()].index
        for nan_index in nan_indices:
            # NaN 값의 날짜 가져오기
            nan_date = group.loc[nan_index, '날짜']
            # 동일 시간대에서 이후 5일 간의 데이터를 찾기
            mask = (new_df['계약번호'] == contract) & (new_df['시간'] == hour) & (new_df['날짜'] > nan_date) & (new_df['날짜'] <= nan_date + pd.Timedelta(days=5))
            future_values = new_df.loc[mask, '전력량']
            # NaN 값이 아닌 경우에만 평균을 계산
            if not future_values.isna().all():
                mean_value = future_values.dropna().mean()
                new_df.loc[nan_index, '전력량'] = mean_value

    return new_df

# 이 함수를 데이터프레임에 적용
df_1_filled2 = fill_nan_with_future_mean(df_1_filled1)

# 결과 확인
df_1_filled2.head()
```

```python
def add_max_and_average_to_df(df):
    # '계약번호'와 '날짜'를 제외한 나머지 열에서 최대값을 구한다
    max_values_per_row = df.drop(['계약번호', '날짜'], axis=1).max(axis=1)
    # 각 행의 최대값을 새로운 열 'row_max'로 추가
    df['row_max'] = max_values_per_row
    # 계약번호별로 그룹화하여 최대값들의 평균을 구한다
    average_max_per_contract = max_values_per_row.groupby(df['계약번호']).transform('mean')
    # 계약번호별 행별 최대값의 평균을 새로운 열 'contract_average_max'로 추가
    df['contract_average_max'] = average_max_per_contract
    return df

# 실제 데이터프레임에 함수 적용
df = add_max_and_average_to_df(df)
```

```python
def calculate_capacity(df):
    # 이상치를 제거한 후 최대값을 저장할 딕셔너리
    capacity_dict = {}

    # 계약번호별로 반복
    for contract_number in df['계약번호'].unique():
        # 계약번호별 데이터 필터링
        contract_data = df[df['계약번호'] == contract_number]
        # IQR 계산
        Q1 = contract_data['row_max'].quantile(0.25)
        Q3 = contract_data['row_max'].quantile(0.75)
        IQR = Q3 - Q1
        # 이상치가 아닌 범위 계산
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        # 이상치가 아닌 값을 필터링
        filtered_data = contract_data[(contract_data['row_max'] >= lower_bound) &
                                      (contract_data['row_max'] <= upper_bound)]
        # 이상치가 제거된 값 중 최대값을 capacity로 설정
        capacity_dict[contract_number] = filtered_data['row_max'].max()

    # capacity 열을 데이터프레임에 추가
    df['capacity'] = df['계약번호'].map(capacity_dict)
    return df

# 실제 데이터프레임에 함수 적용
df = calculate_capacity(df)
```

```python
# 최대값을 capacity로 하여 데이터 치환
def normalize_data(row):
    max_val = row.max()
    return row / max_val

# 시간별 데이터 칼럼 선택
time_columns = [f'{i}시' for i in range(1, 25)]

# 각 계약번호별로 시간대 데이터를 capacity에 비례하도록 정규화
df[time_columns] = df[time_columns].apply(normalize_data, axis=1)

# 클러스터별(기업 크기별)로 데이터 프레임 분할
df_small = df[df['cluster'] == 0]
df_large = df[df['cluster'] == 1]
df_medium = df[df['cluster'] == 2]

# 확인을 위한 출력
print("Small Business Data Frame:")
print(df_small.head())
print("\nMedium Business Data Frame:")
print(df_medium.head())
print("\nLarge Business Data Frame:")
print(df_large.head())
```

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
import matplotlib.pyplot as plt

# 함수 정의: ACF 및 PACF 그래프와 숫자로 나타내기
def acf_pacf_analysis(data, group_name):
    fig, ax = plt.subplots(2, 1, figsize=(10, 8))
    max_lag = min(len(data) // 2, 12)  # 최대 lag 값 설정
    plot_acf(data, ax=ax[0], lags=max_lag)  # 최대 lag 값으로 설정
    plot_pacf(data, ax=ax[1], lags=max_lag)  # 최대 lag 값으로 설정
    plt.title(f"{group_name} ACF 및 PACF")
    plt.show()

    # ACF 및 PACF 수치 출력
    acf_vals = pd.Series(data).autocorr(lag=1)  # ACF 값
    pacf_vals = pd.Series(data).diff().dropna().autocorr(lag=1)  # PACF 값
    print(f"{group_name} ACF(1): {acf_vals}, PACF(1): {pacf_vals}")

# Small Business Data Frame에 대한 ACF 및 PACF 분석
print("Small Business Data Frame ACF 및 PACF 분석 결과:")
data_small = df_small.loc[:, '1시':'24시']  # 시간대 데이터 선택
acf_pacf_analysis(data_small.mean(axis=0), "Small Business")

# Medium Business Data Frame에 대한 ACF 및 PACF 분석
print("\nMedium Business Data Frame ACF 및 PACF 분석 결과:")
data_medium = df_medium.loc[:, '1시':'24시']  # 시간대 데이터 선택
acf_pacf_analysis(data_medium.mean(axis=0), "Medium Business")

# Large Business Data Frame에 대한 ACF 및 PACF 분석
print("\nLarge Business Data Frame ACF 및 PACF 분석 결과:")
data_large = df_large.loc[:, '1시':'24시']  # 시간대 데이터 선택
acf_pacf_analysis(data_large.mean(axis=0), "Large Business")
```
