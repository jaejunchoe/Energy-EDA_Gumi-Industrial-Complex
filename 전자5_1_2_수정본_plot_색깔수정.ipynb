```python
from tslearn.clustering import TimeSeriesKMeans
from tslearn.preprocessing import TimeSeriesScalerMeanVariance
import numpy as np

# df는 이미 준비된 데이터프레임으로 가정합니다.
# 예를 들어, grouped_df가 특정 처리를 거친 시계열 데이터를 포함하고 있다고 가정할 때:
# grouped_df = df.groupby(...).agg(...)

# 시계열 데이터 준비
# '계약번호' 열을 제외한 나머지를 사용하여 시계열 데이터를 준비합니다.
time_series_data = grouped_df.drop('계약번호', axis=1).values
time_series_data = time_series_data.reshape((time_series_data.shape[0], time_series_data.shape[1], 1))

# 데이터 스케일링
scaler = TimeSeriesScalerMeanVariance()
time_series_scaled = scaler.fit_transform(time_series_data)
```

```python
from sklearn.metrics import silhouette_score
```

```python
from sklearn.metrics import silhouette_score

# 실루엣 점수를 위한 클러스터 범위 설정
range_n_clusters = [2, 3, 4, 5, 6, 7, 8, 9, 10]

# 실루엣 점수와 엘보우 방법에 대한 점수를 저장할 리스트
silhouette_scores = []
inertia_scores = []

# 클러스터링을 수행하고 실루엣 점수와 inertia 값을 계산
for n_clusters in range_n_clusters:
    model = TimeSeriesKMeans(n_clusters=n_clusters, metric="dtw", random_state=42)
    labels = model.fit_predict(time_series_scaled)

    silhouette_avg = silhouette_score(time_series_scaled.reshape(time_series_scaled.shape[0], time_series_scaled.shape[1]), labels)
    silhouette_scores.append(silhouette_avg)

    inertia_scores.append(model.inertia_)

# 최대 실루엣 점수에 해당하는 클러스터 수를 찾음
optimal_n_clusters_silhouette = range_n_clusters[silhouette_scores.index(max(silhouette_scores))]
print(f"Optimal number of clusters by silhouette score: {optimal_n_clusters_silhouette}")

# 엘보우 방법: inertia 그래프를 그려 'elbow'를 찾음
plt.figure(figsize=(10, 5))

# Inertia plot
plt.subplot(1, 2, 1)
plt.plot(range_n_clusters, inertia_scores, marker='o')
plt.title('Elbow Method For Optimal k')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')

# Silhouette score plot
plt.subplot(1, 2, 2)
plt.plot(range_n_clusters, silhouette_scores, marker='o')
plt.title('Silhouette Score For Optimal k')
plt.xlabel('Number of clusters')
plt.ylabel('Silhouette score')

plt.tight_layout()
plt.show()

# 엘보우 지점을 정량적으로 찾는 코드는 더 복잡할 수 있으며, 그래프를 보고 결정하는 것이 일반적입니다.
# 그러나 엘보우 지점을 찾는 자동화된 방법도 존재합니다 (예: Kneedle 알고리즘 사용).

# Inertia 값의 변화율 계산
inertia_diff = np.diff(inertia_scores)

# 변화율이 가장 큰 지점의 인덱스를 찾음 (이는 elbow 지점에 해당할 수 있음)
# np.diff는 N-1 길이의 배열을 반환하므로, 인덱스 조정이 필요함
elbow_point_index = np.argmin(inertia_diff) + 1  # +1은 인덱스 조정을 위함

# 실제 클러스터 수에 맞게 조정
optimal_n_clusters_elbow = range_n_clusters[elbow_point_index]

print(f"Optimal number of clusters by elbow method: {optimal_n_clusters_elbow}")
```

```python
from tslearn.clustering import TimeSeriesKMeans
import numpy as np
import pandas as pd

# 가상의 시계열 데이터와 grouped_df를 준비하는 코드는 생략되어 있습니다.
# time_series_scaled에는 적절한 데이터가 들어있다고 가정합니다.
# grouped_df는 이미 계약번호와 다른 필요한 정보를 포함하고 있다고 가정합니다.

n_clusters = 4

# 클러스터링 수행
model = TimeSeriesKMeans(n_clusters=n_clusters, metric="dtw", verbose=True, random_state=42)
labels = model.fit_predict(time_series_scaled)

# 클러스터 레이블을 1부터 시작하게 조정
labels += 1

# 클러스터링 결과를 데이터프레임에 추가
grouped_df['cluster'] = labels

# 결과 출력
print(grouped_df[['계약번호', 'cluster']])
```
